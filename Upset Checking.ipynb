```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

path = 'ncaam-march-mania-2021/MDataFiles_Stage1/'
```

```python
df = pd.read_csv(path+'MRegularSeasonCompactResults.csv')
```

```python
pomeroy = pd.read_csv('Pomeroy.csv')
```

```python
df_features_season = df[df['Season']>2001]
```

```python
df_features_season = pd.merge(
    df_features_season,
    pomeroy,
    how='left',
    left_on=['Season', 'WTeamID'],
    right_on=['Season', 'TeamID']
).rename(columns={
    'Rank':'RankW',
    'AdjEM':'AdjEMW',
    'AdjO':'AdjOW',
    'AdjD':'AdjDW',
    'Luck':'LuckW',
}).drop('TeamID',axis=1)
```

```python
df_features_season
```

```python
df_features_season = pd.merge(
    df_features_season,
    pomeroy,
    how='left',
    left_on=['Season', 'LTeamID'],
    right_on=['Season', 'TeamID']
).rename(columns={
    'Rank':'RankL',
    'AdjEM':'AdjEML',
    'AdjO':'AdjOL',
    'AdjD':'AdjDL',
    'Luck':'LuckL',
}).drop('TeamID',axis=1)
```

```python
df_features_season[df_features_season.isna().any(axis=1)]
```

```python
df_features_season.fillna(0,inplace=True)
```

```python
upsets = df_features_season[df_features_season['RankW'] > df_features_season['RankL']].groupby(['Season','WTeamID']).count().reset_index()[['Season','WTeamID','RankW']]
```

```python
upsets = upsets.rename(columns={'WTeamID':'TeamID','RankW':'Upsets'})
```

```python
upsets.fillna(0,inplace=True)
```

```python
upsets.to_csv("Upsets.csv",index=False)
```

```python
tourney = pd.read_csv('MNCAATourneyCompactResults.csv')
```

```python
seeds = pd.read_csv("MNCAATourneySeeds.csv")
seeds['Seed'] = seeds['Seed'].apply(lambda x: int(x[1:3]))

tourney = tourney.merge(seeds,how="left",left_on=['Season','WTeamID'],right_on=['Season','TeamID']).rename(columns={'Seed': 'SeedW'})
tourney = tourney.merge(seeds,how="left",left_on=['Season','LTeamID'],right_on=['Season','TeamID']).rename(columns={'Seed': 'SeedL'})

```

```python
tourney = tourney.merge(upsets,how="left",left_on=['Season','WTeamID'],right_on=['Season','TeamID']).rename(columns={'Upsets': 'UpsetsW'})
tourney = tourney.merge(upsets,how="left",left_on=['Season','LTeamID'],right_on=['Season','TeamID']).rename(columns={'Upsets': 'UpsetsL'})

tourney =  tourney.T.drop_duplicates().T
```
